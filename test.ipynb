```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.1*np.pi, 2*np.pi, 0.01)
y = np.cos(x) + np.log(x)

plt.figure()
plt.plot(x, y)
plt.title('Default Plot')
```

```python

plt.figure()
plt.plot(x, y, '--', color='#1122FF')
plt.title('Dotted Plot')
plt.legend('line 1')
plt.show()
```

```python

y1 = np.cos(x)
y2 = np.log(x)
y3 = y

plt.figure()
plt.plot(x, y1, ':g', linewidth=3)
plt.plot(x, y2)
plt.plot(x, y3)

legend_names =[]

for i in range(3):
    legend_names.append(f'line {i}')

plt.legend(legend_names)
```
